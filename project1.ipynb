```python
# 요소를 찾는 함수
def get_number_tags():
    return driver.find_elements(By.CSS_SELECTOR, 'div.slick-cell.l8.r8')

# 텍스트 추출 함수
def extract_text():
    number_tags = get_number_tags()  # 요소를 검색
    for i in number_tags:
        text = i.text.strip() if i.text else ""  # i.text가 None일 경우 빈 문자열로 초기화하고, 공백 제거
        if text:  # text가 비어있지 않은 경우에만 출력
            print(text)  # 수정된 텍스트 출력

# 페이지가 로드될 때까지 대기
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.slick-cell.l8.r8')))

# 텍스트 추출 함수 호출
extract_text()
```
